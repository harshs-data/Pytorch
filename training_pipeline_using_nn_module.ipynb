### Train test split

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:], df.iloc[:,0], test_size=0.2, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# scaling

```python
scaler = StandardScaler()
df_train = scaler.fit_transform(X_train)
df_test = scaler.transform(X_test)
```

### LabelEncoder

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

### Numpy to pytorch tensors

```python
X_train_tensor = torch.from_numpy(df_train.astype(np.float32))
X_test_tensor = torch.from_numpy(df_test.astype(np.float32))
y_train_tensor = torch.from_numpy(y_train.astype(np.float32))
y_test_tensor = torch.from_numpy(y_test.astype(np.float32))
```

```python
X_train_tensor.shape
```

### Creating batches

```python
from torch.utils.data import Dataset, DataLoader
```

```python
class CustomDataset(Dataset):

  def __init__(self, features, labels):
    self.features = features
    self.labels = labels

  def __len__(self):
    return len(self.features)

  def __getitem__(self, index):
    return self.features[index], self.labels[index]
```

```python
train_dataset = CustomDataset(X_train_tensor, y_train_tensor)
test_dataset = CustomDataset(X_test_tensor, y_test_tensor)
```

```python
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

### Defining the Model

```python
class MysimpleNN(nn.Module):

  def __init__(self, num_features):

    super().__init__()
    self.linear = nn.Linear(num_features,1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    out = self.linear(features)
    out = self.sigmoid(out)

    return out

  '''
  def loss_function(self, y_pred, y):
    # Clamp predictions to avoid log(0)
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)

    # Calculate loss
    loss = -(y_train_tensor * torch.log(y_pred) + (1 - y_train_tensor) * torch.log(1 - y_pred)).mean()
    return loss
    '''
```

Imp. parameters

```python
learning_rate = 0.1
epochs = 25
```

```python
# defining loss function
loss_function = nn.BCELoss()
```

### Training pipeline

```python
# create model
model = MysimpleNN(X_train_tensor.shape[1])
```

```python
# define optimizer
optimizer = torch.optim.SGD(model.parameters(), lr = learning_rate)
```

```python
# define loop

for epoch in range(epochs):

  for batch_features, batch_labels in train_loader:

    # forward pass
    y_pred = model(X_train_tensor)

    # loss calculate
    loss = loss_function(y_pred, y_train_tensor.view(-1,1)) # Reshape y_train_tensor

    # zero gradients
    '''
    for param in model.parameters():
      if param.grad is not None:
          param.grad.zero_()
    '''
    optimizer.zero_grad()

    # backward pass
    loss.backward()

    # parameters update
    '''
    with torch.no_grad():
      for param in model.parameters():
        param -= learning_rate * param.grad
    '''
    optimizer.step()

    # print loss in each epoch

    print(f'Epoch: {epoch+1}/{epochs}, Loss: {loss.item()}')

```
